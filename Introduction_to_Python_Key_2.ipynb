### Arrays
In many cases, we would love to store some numbers (such as the molecular dynamics data we are dealing with) in a single object (such as the "data" object above). This is easily done by using either arrays or lists. Arrays are better when dealing with numbers, so we'll stick to those for the time being. Luckily enough, numpy is here to help: there are arrays and numpy arrays; the latter are better, because you can use numpy to perform a lot of interesting operations with them in no time. The first thing you have to do with a numpy array is to declare/initialise it:

````python
# If you know how many elements should be contained in your numpy array, say, 10...
array_1 = np.zeros(10)
print(array_1) # always good to print stuff to make sure you are doing the right thing!

# If you don't really know how many elements you want in your array, no problem...
array_2 = np.array([])
print(array_2)
````

So: we have stored in an array object called "array_1" the result of "np.zeros(10)". "np.zeros" calls the function "zeros", one of the very many available when using numpy - which we call "np" in here, remember? If you specify a number (10 in the example above) your array will contain that many elements, all of them initialised to zero. If you don't know how many elements you want your array to be, the syntax is slighty different, and uses the "array" function of numpy to actually transform the empty list [] into a numpy array of unknown length. Ok, now let's populate our array!

````python
array_1[0] = 99
print(array_1)
````

Accessing a single element of a numpy array is easy, you just specify the index you want in between [] brackets - but remember, indexing starts from 0, not 1!

The "data" array we are dealing with, though, has two dimensions, in that it has rows and columns. In that case, you can access the element in column $i$ and row $j$ as [i,j], that is:

````python
data[7,3]
````

This gives you the element in the sixth row and second column. What about accessing a certain range of elements? Say, rows from 4 to 10 within th 1st and 4th column:

````python
data[3:10,[0,3]]
````

Here, we have specified a contiguous range of rows (from 3 to 10) via the ":" operator. Note that the ":" operator alone retrieves all columns or rows. In order to select non contiguous columns (1st and 4th) instead we had to add another [] and separate the indexes via a ",". For instance, to get the first rows and all the columns...

````python
data[0,:]
````

````python
ar_A = np.random.rand(100)
ar_B = np.random.rand(100)
ar_AA = ar_A[0:10]
ar_BB = ar_B[90:]
print(ar_AA)
print(ar_BB)
np.dot(ar_AA,ar_BB)
````

### Histograms vs Probability distribution functions
In many many cases it is also worth looking at how your data are distributed: that is, whether they are homogeneously spread across a certain interval or maybe clustered in some specific regions. The first thing you could do is to build a histogram: divide your range of possible values in "bins", and count how many values fall in each bin. We could do this **by hand**...

````python
# Let's consider the energy values within our "data" object
# Let's find the smallest one and the largest one
energy = data[:,1]
smallest = np.amin(energy)
largest = np.amax(energy)

# Divide the range from smallest to largest into n_bins
r_range = largest - smallest
n_bins = 20
bin_size = r_range/n_bins

# Initialise an array where to store the energy values falling into each bin
histogram = np.zeros((n_bins,2))

# Find the values of energy that fall within each bin
for i in range (0, n_bins): # loop over the bins
    lower_bound = smallest+(i*bin_size)
    upper_bound = lower_bound + bin_size
    histogram[i,0] = lower_bound + (bin_size/2)
    for j in range (0, len(energy)): # (Nested!) loop over the energy values
        if energy[j] >= lower_bound and energy[j] < upper_bound :
            histogram[i,1] = histogram[i,1] +1


# Labeling of x and y axes
plt.xlabel('Energy [kJ/mol]',fontsize=16)
plt.ylabel('Histogram of the Energy',fontsize=16)

# Bigger labels is better!
plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='minor', labelsize=14)

axes = plt.gca()
plt.bar(histogram[:,0],histogram[:,1],width=bin_size)
plt.show()
````

Or, we could use **numpy** (of course!):

````python
h_np = np.histogram(energy, bins=20)
e_np = h_np[0] # values of the energies
b_np = h_np[1] # values of the bins
bb_np = b_np[1:] # for some reason it gets one more
np_bin_size = abs(bb_np[1]-bb_np[0])

# Labeling of x and y axes
plt.xlabel('Energy [kJ/mol]',fontsize=16)
plt.ylabel('Histogram of the Energy',fontsize=16)

# Bigger labels is better!
plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='minor', labelsize=14)
plt.bar(bb_np,e_np,width=np_bin_size)

plt.show()
````

Or, in fact, we could use **matplotlib** directly:

````python
# Labeling of x and y axes
plt.xlabel('Energy [kJ/mol]',fontsize=16)
plt.ylabel('Histogram of the Energy',fontsize=16)

# Bigger labels is better!
plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='minor', labelsize=14)
plt.hist(energy, bins=20)
plt.show()
````

However, these are histograms - *not* probability densities! The latter are characterised by the fact that the total area of the histogram is equal to 1. This is very useful, as it allows you to compare different datasets with different number of data. How do we get these probability densities? Matplotlib can do it...

````python
# Labeling of x and y axes
plt.xlabel('Energy [kJ/mol]',fontsize=16)
plt.ylabel('PDF(Energy)',fontsize=16)

# Bigger labels is better!
plt.tick_params(axis='both', which='major', labelsize=14)
plt.tick_params(axis='both', which='minor', labelsize=14)
plt.hist(energy, bins=20, density="true")
plt.show()
````

The graph looks exactly the same as the histogram - and it is, but look at the y-axis: the heigh of each bin has been normalised in such a way that the total area spanned by all the bins is equal to 1. An even nicer representation featuring a "smoothing" of the bins via what is called a [kernel density extimate](https://en.wikipedia.org/wiki/Kernel_density_estimation) (KDE) is available via e.g. seaborn:

````python
plt.figure(figsize=(6,4), dpi= 80, facecolor='w', edgecolor='k') 
plt.tick_params(axis='both', which='major', labelsize=20)
plt.tick_params(axis='both', which='minor', labelsize=20)
sns.distplot(energy, bins=20);
plt.xlabel('Energy [kJ/mol]', fontsize=16)
plt.ylabel('PDF(Energy)', fontsize=16)
plt.show()
````
